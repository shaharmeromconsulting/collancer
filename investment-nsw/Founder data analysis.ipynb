# 2 Define functions

```python
def share_calculator(data_set,columns,numerator_filter,denominator_filter,new_index,new_col):
    numerator = data_set.loc[numerator_filter,:].reset_index(drop=True)
    denominator = data_set.loc[denominator_filter,:].reset_index(drop=True)
    numerator = numerator.set_index(new_index)
    denominator = denominator.set_index(new_index)
    numerator[columns] = numerator[columns]/denominator[columns]
    for key,val in new_col.items(): numerator[key] = val
    return numerator[[key]+columns]

def region_cleaner(_table):
    _output = _table.copy()
    _output['region'] = _output['region'].astype(str)
    _output['region'] = _output['region'].apply(lambda x: x.replace('\xa0',' ').replace('\xad','').strip())
    _output['region'] = _output['region'].apply(lambda x: x.replace(', United States', ''))
    _output['region'] = _output['region'].apply(lambda x: x.replace(', Canada', ''))
    _output['region'] = _output['region'].apply(lambda x: x.replace(', Australia', ''))
    return _output

def year_summer(_table):
    _output = _table.copy()
    #years in data
    year_cols = [str(year) for year in range(2000,2025)]
    for y in year_cols:
        try:_output[y] = _output[y].apply(lambda x: float(x.replace(',','')))
        except: pass
    _output['2000_2005'] = _output[[str(year) for year in range(2000,2005)]].apply(lambda x: sum(x),axis=1)
    _output['2005_2010'] = _output[[str(year) for year in range(2005,2010)]].apply(lambda x: sum(x),axis=1)
    _output['2010_2015'] = _output[[str(year) for year in range(2010,2015)]].apply(lambda x: sum(x),axis=1)
    _output['2015_2020'] = _output[[str(year) for year in range(2015,2020)]].apply(lambda x: sum(x),axis=1)
    _output['2020_2025'] = _output[[str(year) for year in range(2020,2025)]].apply(lambda x: sum(x),axis=1)
    _output['2000_2024'] = _output[year_cols].apply(lambda x: sum(x),axis=1)
    return _output
```

```python
year_cols = [str(x) for x in range(2000,2025)]

founder_data = pd.read_csv('master_data.csv').iloc[:,1:]
founder_data = region_cleaner(founder_data)
founder_data = year_summer(founder_data)


year_cols = year_cols+['2000_2005','2005_2010','2010_2015','2015_2020','2020_2025','2000_2024',]
founder_data
```
